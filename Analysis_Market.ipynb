```python
import pandas as pd

file = pd.read_excel("./Data_Coffee_Market_Revenue.xlsx", sheet_name=1)
file = file.drop(index=[0,1,2,3])  
file = file.drop(columns=['Unnamed: 0'])
file.columns = ["Year", "Instant", "Roast"]
file["Instant"] = pd.to_numeric(file["Instant"], errors="coerce")
file["Roast"] = pd.to_numeric(file["Roast"], errors="coerce")
file.head(5)
```

```python
import matplotlib.pyplot as plt

# Create the line plot
plt.figure(figsize=(15, 5))
plt.plot(file["Year"], file["Instant"], marker="o", label="Instant Coffee")
plt.plot(file["Year"], file["Roast"], marker="o", label="Roast Coffee")

plt.title("Coffee Market Revenue in the UK (2018–2030)")
plt.xlabel("Year")
plt.ylabel("Revenue (billion GBP)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

The revenue is forecast to experience significant growth in all segments in 2030. Comparing the two different segments for that year, the segment 'Roast Coffee' leads the ranking with 5.56 billion British pounds. Contrastingly, 'Instant Coffee' ranks below, with 4.63 billion British pounds. 

Data provided by Statista Market Insights are estimates. The Coffee segment consists of Roast Coffee, that incorporates caffeinated and decaffeinated coffee in the form of ground coffee and whole beans, that is prepared with coffee machines or a French press, as well as Instant Coffee, that includes soluble coffee that does not need further equipment for preparation.
While Roast Coffee, or fresh coffee is especially popular in traditionally coffee-leaning countries like the U.S., Germany, Italy or Brazil, Instant Coffee is important in traditionally tea-leaning countries like the UK, Czechia or Asia in general as the method of preparation is very similar to that of tea. 

```python
file1 = pd.read_excel("./Data_Coffee_Stores.xlsx", sheet_name=0)
file1 = file1.drop(index=[0,1,2,3])  
file1 = file1.drop(columns=['Unnamed: 0'])
file1.columns = ["Company", "2021", "2022","2023","2024","2025"]
years = ["2021", "2022", "2023", "2024", "2025"]
file1_long = file1.melt(id_vars="Company", value_vars=years,
                  var_name="Year", value_name="Stores")
file1_long.head(5)
```

```python
import numpy as np

plt.figure(figsize=(12, 6))
pivot = file1_long.pivot(index="Year", columns="Company", values="Stores")

pivot.plot(kind="bar", figsize=(10, 5))
plt.title("Stores per Company - Leading Coffee Shop Chains in the UK")
plt.xlabel("Year")
plt.ylabel("Stores")
plt.legend()
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```

Published by Allegra World Coffee Portal in Jan/2025

As of January 2025, Costa Coffee had the most units among the leading coffee shop chains in the United Kingdom, with 2,671 outlets. Greggs and Starbucks ranked second and third, respectively, with 2,610 and 1,354 outlets. 

 
 What are the leading UK coffee shop chains by revenue? 
 As well as having the most units across leading coffee shop chains, Costa Coffee was the leading coffee shop chain with the highest revenue in the UK in 2023, with the company generating over one billion British pounds in revenue that year. Meanwhile, Costa Coffee's competitors, Starbucks and Caffè Nero, both generated less than 600 million British pounds in revenue in 2023.  
 How popular is Starbucks in the UK? 
 Starbucks is the only non-British company in the ranking. Although the company’s popularity failed to match the success of Costa Coffee in 2023, the  number of Starbucks stores in the UK was higher than in any other European country. The overall number of Starbucks stores in the UK has grown gradually since 2012, with the company from Seattle seeing increases in both licensed stores and company-operated stores since 2020.
